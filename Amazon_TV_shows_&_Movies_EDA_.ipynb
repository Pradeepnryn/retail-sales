THe column "Description" is dropped from the dataset as it has got no relevance with the objective of the project ie..focusing on Content Diversity,Regional Availability,Trends Over Time,IMDb Ratings & Popularity

````python
#Fillng Missing Values

final_data['age_certification'].fillna('Not Rated',inplace=True) #Replace missing values for the age_certification column=Not rated
````

The age_certification column contains information about the content rating assigned to movies and shows. Since this variable is categorical, missing values cannot be replaced using statistical measures such as mean or median. Therefore, the missing values were replaced with the label “Not Rated” to indicate that the rating information is not available. This approach preserves the column for further analysis while clearly distinguishing records with unavailable certification data.

````python
#skewness  of Numerical columns with Missing values

print("skewness of column imdb_score:",final_data['imdb_score'].skew())
print("skewness of column imdb_votes:",final_data['imdb_votes'].skew())
print("skewness of column tmdb_popularity:",final_data['tmdb_popularity'].skew())
print("skewness of column tmdb_score :",final_data['tmdb_score'].skew())


````

In this step, we calculated the skewness of important numerical columns such as IMDb score, IMDb votes, TMDB popularity, and TMDB score to understand the distribution of the data. Skewness helps identify whether the data is symmetrically distributed or biased toward higher or lower values. Performing this analysis is important because it guides the selection of appropriate techniques for handling missing values. Based on the skewness results, suitable statistical measures such as the median can be used to fill missing values, ensuring that the data distribution is not significantly affected by outliers

````python
#Filling missing values of numerical columns

# imdb_score (use mean)
final_data['imdb_score'].fillna(final_data['imdb_score'].mean(), inplace=True)

# imdb_votes (use median)
final_data['imdb_votes'].fillna(final_data['imdb_votes'].median(), inplace=True)

# tmdb_popularity (use median)
final_data['tmdb_popularity'].fillna(final_data['tmdb_popularity'].median(), inplace=True)

# tmdb_score (use median)
final_data['tmdb_score'].fillna(final_data['tmdb_score'].median(), inplace=True)
````

Based on the skewness analysis of numerical variables, appropriate imputation techniques were selected for handling missing values. The IMDb score variable showed a nearly symmetric distribution, therefore the mean was used to replace missing values. However, variables such as IMDb votes, TMDB popularity, and TMDB score exhibited skewed distributions, so the median was used for imputation to minimize the influence of extreme values.

The person_id column represents a unique identifier for individuals associated with the titles. Since it serves only as an identifier and does not represent a measurable numerical variable, statistical imputation methods such as mean or median are not appropriate for this column. Therefore, the missing values in person_id were retained (or can be replaced with a placeholder value) as they do not significantly affect the analytical objectives of the study.

````python
#Filling missing values of categorical columns

# Name (use mode)
final_data['name'].fillna(final_data['name'].mode()[0], inplace=True)

# Character (use mode)
final_data['character'].fillna(final_data['character'].mode()[0], inplace=True)

# role (use mode)
final_data['role'].fillna(final_data['role'].mode()[0], inplace=True)

````

Missing values in categorical columns such as name, character, and role were handled using the mode, which represents the most frequently occurring value in each column. Since these variables contain categorical information related to cast and roles, using the mode helps preserve the most common category in the dataset while maintaining the overall distribution of the data.

For the imdb_id column, which represents the unique identifier assigned to titles in the IMDb database, statistical imputation methods such as mean, median, or mode are not appropriate because it functions only as an identifier. Therefore, the missing values were retained or replaced with a placeholder value, as they do not significantly impact the analytical objectives of this study.

````python
#cleaning the genre column

final_data['genres'] = final_data['genres'].str.replace('[','', regex=False)\
                                           .str.replace(']','', regex=False)\
                                           .str.replace("'","")
final_data['genres'] = final_data['genres'].str.split(',')
final_data['genres']
````

The genres column contained values stored as strings with brackets and quotation marks. Using the .str.replace() function, these unwanted characters were removed from the column. The .str accessor allows string operations to be applied to each value in the column, while replace() removes specific characters. This cleaning step makes the genre values easier to interpret and analyze during exploratory data analysis.

The genres column originally contained multiple genres stored in a single string separated by commas. To make the data easier to analyze, the column was split into a list of individual genres using the str.split() function. This allows each genre to be handled separately during further analysis and visualization.

````python
final_data['genres'].value_counts().head(10)
````

In the above output we see same combination of genres presented as two differnet combinations with change in order ...eg:[drama ,romance] and [romance ,drama].we need to sort this

````python
# Standardizing the order of genres to avoid duplicate combinations #challenge

final_data['genres'] = final_data['genres'].apply(
    lambda x: sorted([g.strip() for g in x]) if isinstance(x, list) else x
)

# convert to string
final_data['genres'] = final_data['genres'].apply(
    lambda x: ', '.join(x) if isinstance(x, list) else x
)
````

In this step, the genres column was standardized to ensure accurate analysis of genre combinations. The column originally contained multiple genres in list format, where the same combinations appeared in different orders (for example, “drama, comedy” and “comedy, drama”). These variations would be treated as different categories during analysis, leading to incorrect frequency counts.

To resolve this issue, each list of genres was processed using the apply() function. A lambda function was used to iterate through each row, where extra spaces were removed using the strip() function. The genres were then sorted alphabetically using the sorted() function to ensure a consistent order across all records.

After standardizing the order, the list of genres was converted back into a string format using the join() function. This transformation ensures that identical genre combinations are represented uniformly, allowing accurate counting and meaningful visualization.

Overall, this step improves data quality by eliminating inconsistencies in genre representation and ensures reliable insights during exploratory data analysis.

````python
#cleaning the production_countries column

final_data['production_countries'] = final_data['production_countries'].str.replace('[','', regex=False)\
                                                                       .str.replace(']','', regex=False)\
                                                                       .str.replace("'","")
final_data['production_countries'] = final_data['production_countries'].str.split(',')
final_data['production_countries'] = final_data['production_countries'].apply(
    lambda x: [i.strip() for i in x]
)
````
